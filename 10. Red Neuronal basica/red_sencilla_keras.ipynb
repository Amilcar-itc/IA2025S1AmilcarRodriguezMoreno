```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Se entrena bajo la funcion cuadratica:
# y = 2x² - 2x + 3

# La siguiente funcion devuelve el valor de y al ingresar x en la funcion
def funcion_objetivo(x):
    return 2 * x**2 - 2 * x + 3


x_train = np.linspace(-5, 5, 100)
y_train = funcion_objetivo(x_train)

# Red neuronal
modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu', input_shape=[1]),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1)  # salida
])

modelo.compile(optimizer='adam', loss='mse')

# Entrenamiento
modelo.fit(x_train, y_train, epochs=200, verbose=0)

x_test = np.linspace(-6, 6, 200)
y_pred = modelo.predict(x_test)

# Graficacion
plt.plot(x_train, y_train, label='Función Real')
plt.plot(x_test, y_pred, label='Predicción Red Neuronal', linestyle='dashed')
plt.legend()
plt.title("Aproximación de la función cuadrática")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
